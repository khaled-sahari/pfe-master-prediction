```python
import pandas as pd
import joblib
import json
from tensorflow.keras.models import load_model
```

```python
# LOAD CONFIG
with open("models/config.json") as f:
    config = json.load(f)

features = config["features"]
targets = config["targets"]
```

```python
# LOAD DATA
df = pd.read_csv("FINAL_DATASET_PREDICTION_CEjour.csv")
```

```python
# SIMULER PANNE
df_virtual = df.copy()

df_virtual["W_GAS_real"] = df_virtual["W_GAS"]
df_virtual["WATER_real"] = df_virtual["WATER"]
df_virtual["COND_real"] = df_virtual["COND_VOL"]



X = df_virtual[features]
```

```python
# LOAD MODEL (choix)
model = joblib.load("models/RandomForest.pkl")  # ou LR.pkl
```

```python
# PREDICTION
y_pred = model.predict(X)

df_virtual["GAS_pred"] = y_pred[:,0]
df_virtual["WATER_pred"] = y_pred[:,1]
df_virtual["COND_pred"] = y_pred[:,2]

df_virtual.to_csv("results/virtual_sensor.csv", index=False)

print("✅ Virtual sensor done")
```
