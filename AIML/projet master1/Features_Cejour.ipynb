```python
import pandas as pd

df = pd.read_csv("step9_final_dataset.csv")

df["DATE"] = pd.to_datetime(df["DATE"])

df = df.sort_values(["WELL","DATE"]).reset_index(drop=True)
```

```python
wellnames = df["WELL"].unique()
wellnames
```

```python
lags = [1,3,7]

for lag in lags:
    df[f"WHP_LAG_{lag}"] = df.groupby("WELL")["WHP"].shift(lag)
    df[f"WHT_LAG_{lag}"] = df.groupby("WELL")["WHT"].shift(lag)
    df[f"GAS_LAG_{lag}"] = df.groupby("WELL")["W_GAS"].shift(lag)
    df[f"WATER_LAG_{lag}"] = df.groupby("WELL")["WATER"].shift(lag)
    df[f"COND_LAG_{lag}"] = df.groupby("WELL")["COND_VOL"].shift(lag)
```

```python
for col in ["W_GAS", "WATER", "COND_VOL"]:

    df[f"{col}_MEAN_3"] = (
        df.groupby("WELL")[col]
        .transform(lambda x: x.shift(1).rolling(3).mean())
    )

    df[f"{col}_MEAN_7"] = (
        df.groupby("WELL")[col]
        .transform(lambda x: x.shift(1).rolling(7).mean())
    )
```

```python
df["DAY"] = df["DATE"].dt.day
df["MONTH"] = df["DATE"].dt.month
df["DAY_OF_WEEK"] = df["DATE"].dt.dayofweek
```

```python
df_model = df.dropna().copy()

print(df_model.shape)
```

```python
df.to_csv("FINAL_DATASET_PREDICTION_CEjour.csv", index=False)
```
