```python
import pandas as pd
```

Data preparation

```python
df = pd.read_csv('data.csv')
```

```python
# 2. Nettoyage des colonnes
# Nous ne gardons que ce que vous avez spécifié
cols_to_keep = ['WELL', 'DATE', 'HOURS', 'WHP', 'WHT', 'W_GAS']
df_cleaned = df[cols_to_keep].copy()
```

```python
# 3. Formatage des Dates
# On s'assure que Python reconnaît les dates pour pouvoir les trier correctement
df_cleaned['DATE'] = pd.to_datetime(df_cleaned['DATE'])
```

```python
# 4. Alignement "Puits par Puits" et "Date par Date"
# On trie d'abord par nom de puits (WELL), puis par chronologie (DATE)
df_cleaned = df_cleaned.sort_values(by=['WELL', 'DATE'])
```

```python
# 5. Filtrage des arrêts (Shut-in)
# Si HOURS est 0, il n'y a pas de dynamique d'écoulement à apprendre
df_cleaned = df_cleaned[df_cleaned['HOURS'] > 0]

# 6. Réinitialisation de l'index
df_cleaned = df_cleaned.reset_index(drop=True)
```

```python
# Affichage du résultat pour vérification
print("Données alignées et nettoyées :")
print(df_cleaned.head(10))

# Sauvegarde pour l'étape suivante (ML)
df_cleaned.to_csv('prepared_data_vfm.csv', index=False)
```

```python
# 1. Vérification des valeurs manquantes (NaN)
print("--- Vérification des NaNs ---")
print(df_cleaned.isna().sum())

# 2. Description statistique
print("\n--- Description Statistique ---")
print(df_cleaned.describe())
```
